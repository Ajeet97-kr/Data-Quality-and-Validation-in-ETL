Question 1 : Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?

Data Quality in ETL is the comprehensive practice of ensuring data is accurate, complete, and reliable throughout its journey from source to destination. It transcends data cleaning because, while cleaning focuses on fixing specific errors (like null values or formatting), Data Quality is a proactive strategy. It encompasses broader dimensions such as data governance, architectural integrity, and timeliness, ensuring that the resulting insights are not just "fixed," but are strategically sound and trustworthy for business use.

Question 2 : Explain why poor data quality leads to misleading dashboards and incorrect decisions.

Poor data quality transforms dashboards from "decision-support tools" into "misinformation tools." Because dashboards aggregate data, they can mask underlying errors—like missing values or duplicates—under beautiful visualizations. This creates a false sense of security, leading executives to make high-stakes pivots based on "hallucinated" trends, ultimately resulting in lost revenue, wasted resources, and a total loss of organizational trust in data-driven culture.

Question 3 : What is duplicate data? Explain three causes in ETL pipelines.

Duplicate data is the redundant repetition of records within a dataset, which can lead to overestimation of KPIs and wasted storage. In ETL pipelines, duplicates are primarily caused by non-idempotent retries (re-inserting data after a crash), improper SQL joins (creating a fan-out effect where rows multiply), and source system overlap (extracting the same entity from different departments without a unified identifier).

Question 4 : Differentiate between exact, partial, and fuzzy duplicates.

1. Exact Duplicates
These are "mirror images." Every single column and value in one row is identical to another row.

It happens Usually a technical glitch in the ETL pipeline, such as a job running twice without a "truncate" or "merge" command.

2. Partial Duplicates
These records share the same unique identifier (like a Primary Key or Email), but other attributes in the row differ.

This often occurs when a record is updated over time, but the old version isn't deleted, or when the same customer is entered into two different systems.

3. Fuzzy Duplicates
These are the trickiest to find. The records refer to the same real-world entity, but the data is entered differently due to typos, formatting, or nicknames. There is no shared unique ID.

Human error during manual data entry or lack of data validation at the source causes fuzzy duplicates.



Question 5 : Why should data validation be performed during transformation rather than after loading?

Performing data validation during transformation is a proactive approach that ensures only "clean" data enters the final storage system. It prevents the pollution of the data warehouse, reduces operational costs by avoiding expensive post-load fixes, and ensures referential integrity. By validating early, data engineers create a "gatekeeper" effect that protects downstream users from making decisions based on faulty or incomplete information.

Question 6 : Explain how business rules help in validating data accuracy. Give an example.

Business rules are the "logical filters" that translate real-world organizational requirements into data constraints. They ensure accuracy by checking not just the format of the data, but its plausibility and relationship to other data points. By enforcing rules like "age must be between 0 and 60" or "total must equal subtotal plus tax," organizations prevent logically impossible data from corrupting their decision-making process.

Example: E-commerce Returns
Imagine you are building an ETL pipeline for a clothing retailer.

The Data Point: A customer returns an item.

The Business Rule: "The Refund_Amount must be less than or equal to the Original_Purchase_Price minus any applied discounts."

The Validation Check: If the ETL process encounters a record where a customer paid 50 rupees but is being refunded 75 rupees, the business rule flags this as inaccurate.

Question 7 : Write an SQL query on sales transactions to list all duplicate keys and their counts using the
business key (Customer_ID + Product_ID + Txn_Date + Txn_Amount )

```python
SELECT
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount,
    COUNT(*) AS Duplicate_Count
FROM Sales_Transactions
GROUP BY
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount
HAVING COUNT(*) > 1;
```

Question 8 : Enforcing Referential Integrity

Assume the following Customers_master table:

Identify Sales_Transactions.Customer_ID values that violate referential integrity when joined with
Customers_Master and write a query to detect such violations.

```python
SELECT
    s.Txn_ID,
    s.Customer_ID,
    s.Customer_Name
FROM Sales_Transactions s
LEFT JOIN Customers_Master m ON s.Customer_ID = m.CustomerID
WHERE m.CustomerID IS NULL;
```
